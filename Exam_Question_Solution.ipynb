**Question 01:**

```python
class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner
        self.balance = balance

    def __str__(self):
        return f"Account[Owner: {self.owner} | Balance: ${self.balance:.2f}]"

    def deposit(self, amount):
        if amount <= 0:
            raise ValueError("Deposit amount must be positive")
        self.balance += amount

    def withdraw(self, amount):
        if amount <= 0:
            raise ValueError("Withdraw amount must be positive")
        if self.balance - amount < 0:
            raise ValueError("Insufficient balance")
        self.balance -= amount


def transfer(source_account, target_account, amount):
    source_account.withdraw(amount)
    target_account.deposit(amount)


# Example Usage
acc1 = BankAccount('Alice', 1500)
acc2 = BankAccount('Bob', 500)

print(acc1)
transfer(acc1, acc2, 300)
print(acc1)
print(acc2)
```

**Question 02:**

```python
class Person:
    def __init__(self, name, age):
        self.__name = name
        self.__age = age

    def get_name(self):
        return self.__name

    def get_age(self):
        return self.__age

    def introduce(self):
        print(f"I am {self.__name}, aged {self.__age}.")


class Student(Person):
    def __init__(self, name, age, student_id):
        super().__init__(name, age)
        self.student_id = student_id

    def introduce(self):
        print(f"I am {self.get_name()}, aged {self.get_age()}. Student ID: {self.student_id}")

    def study(self, subject):
        print(f"{self.get_name()} is studying {subject}.")


class Teacher(Person):
    def __init__(self, name, age, salary):
        super().__init__(name, age)
        self.__salary = salary

    def get_salary(self):
        return self.__salary

    def introduce(self):
        print(f"I am {self.get_name()}, aged {self.get_age()}. Salary: ${self.__salary}")


# Example Usage
s = Student('Rafi', 21, 'CSE-2023-45')
t = Teacher('Dr. Hasan', 45, 95000)

s.introduce()
s.study('Python')
t.introduce()
```

```python
class Shape:
    def area(self):
        return 0

    def describe(self):
        print(f"This is a shape with area: {self.area():.2f}")


class Circle(Shape):
    class Validator:
        @staticmethod
        def is_valid(radius):
            return radius > 0

    def __init__(self, radius):
        if not Circle.Validator.is_valid(radius):
            raise ValueError("Invalid radius")
        self.radius = radius

    def area(self):
        return 3.14159 * self.radius * self.radius


class Rectangle(Shape):
    class Validator:
        @staticmethod
        def is_valid(w, h):
            return w > 0 and h > 0

    def __init__(self, width, height):
        if not Rectangle.Validator.is_valid(width, height):
            raise ValueError("Invalid dimensions")
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height


def print_all_areas(shapes):
    for shape in shapes:
        shape.describe()

shapes = [Circle(7), Rectangle(4, 5)]
print_all_areas(shapes)
```
