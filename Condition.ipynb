```python
raihan=int(input("Enter Raihan age: "))
tanvir=int(input("Enter Tanvir age: "))
if raihan>tanvir:
    print("Raihan is older")
else:
    print("Tanvir is older")
```

```python
raihan=int(input("Enter Raihan age: "))
tanvir=int(input("Enter Tanvir age: "))
nyeem=int(input("Enter Nyeem age: "))

if raihan>tanvir and raihan>nyeem:
    print("Raihan is older")
elif tanvir>nyeem:
    print("Tanvir is older")
else:
    print("Nyeem is older")
```

```python
num=int(input("Input number: "))

if num>0:
    print(f"{num} is positive")
elif num<0:
    print(f"{num} is negative")
else:
    print(f"{num} is zero")
```
