Sl No- 10 # **Days conversion formula**

```python
days=int(input("Enter days: "))
years = days//365
weeks = (days - (years*365))//7
remianing_days = days - ((years*365) + (weeks*7))
print(f"\n{days} days = {years} year/s, {weeks} week/s and {remianing_days} day/s")
```
