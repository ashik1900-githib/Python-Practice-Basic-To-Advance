## Python Arithmetic Operators

### Addition

```python
x= 50
y= 100
print("This is addition:",x+y)
```

### Subtraction

```python
x= 100
y= 50
print("This is Subtraction:",x-y)
```

### Multiplication

```python
x= 100
y= 50
print("This is Multiplication:",x*y)
```

### Division

```python
x= 100
y= 50
print("This is Division:",x/y)
```

### Floor division

```python
x= 100
y= 50
print("This is Multiplication:",x//y)
```

### Power (Exponentiation)

```python
x= 10
y= 2
print("This is Power (Exponentiation):",x**y)
```

### Modulus

```python
x= 10
y= 3
print("This is Modulus:",x%y)
```
