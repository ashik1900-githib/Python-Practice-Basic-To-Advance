## Python Loops

#### Python has two primitive loop commands:
    1. While Loops
    2. For Loops

### Python While Loops

```python
x=5
y=1
while x>=y:
    print("The output is:",x)
    x-=1
```

```python
x=0
while x<=10:
    if x%2==0:
        print(x)
    x+=1
```

```python
x=1
while x<=5:
    print("The output is:",x)
    x+=1
```

### The break Statement

```python
i = 1
while i < 7:
  print(i)
  if i == 4:
    break
  i += 1
```

### The continue Statement

```python
i = 0
while i < 5:
  i += 1
  if i == 3:
    continue
  print(i)
```

### The else Statement

```python
i = 1
while i < 6:
  print(i)
  i += 1
else:
  print("i is no longer less than 6")
```

### Even Number (SUM)

```python
x=0
s=0
while x<=10:
    if x%2==0:
        s=s+x
        print(x)
    x=x+1
print(s)
```

```python
x = 1
while x < 10:
  if x==7 or x==9:
      x=x+1
      continue
  print(x)
  x=x+1
```

### Python For Loops

```python
fruits = ["apple", "banana", "cherry"]
for x in fruits:
  print(x)
```

```python
for x in "banana":
  print(x)
```

### The break Statement

```python
fruits = ["apple", "banana", "cherry"]
for x in fruits:
  print(x)
  if x == "banana":
    break
```

```python
fruits = ["apple", "banana", "cherry"]
for x in fruits:
  if x == "banana":
    break
  print(x)
```

### The continue Statement

```python
fruits = ["apple", "banana", "cherry"]
for x in fruits:
  if x == "banana":
    continue
  print(x)
```

### The range() Function

```python
for x in range(6):
  print(x)
```

```python
for x in range(2, 6):
  print(x)
```

```python
for x in range(2, 30, 3):
  print(x)
```

### Else in For Loop

```python
for x in range(6):
  print(x)
else:
  print("Finally finished!")
```

```python
for x in range(6):
  if x == 3: break
  print(x)
else:
  print("Finally finished!")
```

### Nested Loops

```python
adj = ["red", "big", "tasty"]
fruits = ["apple", "banana", "cherry"]

for x in adj:
  for y in fruits:
    print(x, y)
```
